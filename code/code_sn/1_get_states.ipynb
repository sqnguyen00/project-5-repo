```python
#imports
from geopy.geocoders import Nominatim
import pandas as pd
```

```python
#load state data
city_info = pd.read_csv('./data/city_info.csv')
```

```python
city_info.head()
```

```python
#geopy takes in string lat and lon so I am creating columns that will allow me to add string lat and lon easily.
city_info['str_lat'] = city_info['Lat'].astype(str)
```

```python
city_info['str_lon'] = city_info['Lon'].astype(str)
```

```python
#Creating a column "State" that will replace the string of lat and lon with it's respective state. 
city_info['State'] = city_info['str_lat']+','+city_info['str_lon']
```

```python
city_info
```

```python
city_info.drop(columns=['Unnamed: 0', 'str_lat', 'str_lon'], inplace=True)
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
city_info
```

```python
#creating a list of dictionaries where key is the string location and value is the state.
locations = []
for i in city_info['State']:
     location = geolocator.reverse(i)
     #create dictionary from geolocator data
     d = location.raw['address']
     #getting state names from dictionary
     locations.append({i: d.get('state')})
```

```python
#using the list of dictionarys to create one dictionary. 
state = {}
for i in locations:
    state.update(i)
```

```python
#replacing string of location with state
city_info['State'].replace(state, inplace=True)
```

```python
#checking to see if locations have states. we did it!
city_info.head()
```

```python
#saving updated city info.
city_info.to_csv('./data/city_info_state.csv')
```
