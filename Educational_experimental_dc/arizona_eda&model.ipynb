```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import os

from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression,LogisticRegressionCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('./temperature and precipitation/arizona_drought.csv')
```

```python
df.head()
```

```python
df.drop(columns='StatisticFormatID',inplace=True)
df.head()
```

```python
df.drop(columns='ValidEnd',inplace=True)
df.head()
```

```python
df=pd.get_dummies(df,columns=['Region'])
df.head()
```

```python
df=pd.get_dummies(df, columns=['ClimateDivisionsID'])
df.head()
```

```python
df.isnull().sum()
```

```python
df.rename(columns= lambda x:x.replace(' ', '_').lower(), inplace= True)

df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
object_features=['none','d0','d1','d2','d3','d4']
```

```python
df.set_index('validstart', inplace=True)
df.index = pd.to_datetime(df.index)
```

```python
for col in df:
    if df[col].dtype==object:
        df[col]=[float(str(i).replace(",",""))for i in df[col]]
#https://stackoverflow.com/questions/8420143/valueerror-could-not-convert-string-to-float-id     
```

```python
for col in df:
    if df[col].dtype==object:
        df[col]=[float(str(i).replace("''",""))for i in df[col]]
```

```python
df.dtypes
```

```python
df['is_drought_area']=np.where(df['none']==0, 1,0)

df.head()
```

```python
df.drop(columns='mapdate',inplace=True)
```

### Modeling

```python
df['is_drought_area'].value_counts(normalize=True)
```

```python
X=df.drop(columns='is_drought_area')
y=df['is_drought_area']

X_train, X_test, y_train, y_test=train_test_split(X,y, stratify=y, random_state=42)


pipe_1=Pipeline([
    ('ss', StandardScaler()),
    ('logreg', LogisticRegression(solver='liblinear'))
])


pipe_1.fit(X_train,y_train)
```

```python
pipe_1.score(X_train,y_train), pipe_1.score(X_test,y_test)
```

```python
pipe_2=Pipeline([
    ('ss', StandardScaler()),
    ('rf', RandomForestClassifier())
])

pipe_2.fit(X_train,y_train)
```

```python
pipe_2.score(X_train, y_train), pipe_2.score(X_test,y_test)
```
