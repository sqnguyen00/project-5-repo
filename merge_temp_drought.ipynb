```python
temperatures = pd.concat(temp_data)
```

```python
features = ['Unnamed: 0', 'tmax', 'tmin', 'temp_id']
temps = temperatures[features]
prcp_feat = ['Unnamed: 0', 'prcp', 'temp_id']
prcp = temperatures[prcp_feat]
```

```python
temps = temps[temps['Unnamed: 0']<'2020-12-31']
prcp = prcp[prcp['Unnamed: 0']<'2020-12-31']
```

```python
temps = temps[temps['Unnamed: 0']>'2000-12-31']
prcp = prcp[prcp['Unnamed: 0']>'2000-12-31']
```

```python
temps.set_index(pd.to_datetime(temps['Unnamed: 0']), inplace = True)
prcp.set_index(pd.to_datetime(prcp['Unnamed: 0']), inplace = True)
```

```python
temps.index.rename('ValidEnd', inplace=True)
prcp.index.rename('ValidEnd', inplace=True)
```

```python
states_temp = []
states_prcp = []
for temp_id, state in temps.groupby('temp_id'):
    states_temp.append(state.resample('W-MON').mean())
for temp_id, state in prcp.groupby('temp_id'):
    states_prcp.append(state.resample('W-MON').sum())
```
